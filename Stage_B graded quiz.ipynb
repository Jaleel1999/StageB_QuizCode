```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
normalised_df = df.drop(columns=["date","lights"],inplace=True)
normalised_df = pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
appliances_target = normalised_df["Appliances"]
```

```python
normalised_df
```

```python
appliances_target
```

```python
#Splitting the data set with radom_state = 42.
X1_train,X1_test,y1_train,y1_test = train_test_split(normalised_df,appliances_target,test_size=0.3,random_state=42)
```

```python
#Performing Multiple linear regression on normalized data.
regressor1 = LinearRegression()
regressor1.fit(X1_train,y1_train)
```

```python
# Predicting the Test set results
y1_pred = regressor1.predict(X1_test)
```

```python
# importing Mean absolute error library.
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y1_test, y1_pred)
```

```python
mae
```

```python
# Residual sum of squares.
rss = np.sum(np.square(y1_test - y1_pred))
```

```python
rss
```

```python
# Root mean squared error.
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y1_test,y1_pred))
```

```python
rmse
```

```python
# R - Squared or Coefficient of Determination.
from sklearn.metrics import r2_score
r2_score = r2_score(y1_test, y1_pred)
r2_score
```

```python
# Importing and using Ridge Regression.
from sklearn.linear_model import Ridge
ridge_rg = Ridge(alpha=0.4)
ridge_rg.fit(X1_train,y1_train)
```

```python
# Importing and using Lasso Regression.
from sklearn.linear_model import Lasso
ls = Lasso(alpha=0.001)
ls.fit(X1_train,y1_train)
```

```python
def get_weights(model,feat,col_name):
    weights = pd.Series(model.coef_,feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features',col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_weights_df = get_weights(regressor1,X1_train,'Linear_model_weights')
ridge_weights_df = get_weights(ridge_rg,X1_train,'Ridge_weights')
lasso_weights_df = get_weights(ls,X1_train,'Lasso_weights')
```

```python
final_weights = pd.merge(linear_weights_df,ridge_weights_df,on='Features')
final_weights = pd.merge(final_weights,lasso_weights_df,on='Features')
```
